GridSearchCV with the single Estimator :

```python
from sklearn.model_selection import GridSearchCV
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
import pandas as pd

data = load_iris()
x = data.data
y = data.target

param_grid = {"criterion":["gini","entropy"],
               "max_depth":[2,3,4,5,6,7]
              }

result = GridSearchCV(DecisionTreeClassifier(),param_grid,cv=5,return_train_score=False)

result.fit(x,y)
```

```python
print(result.best_params_)         # Best hyperparameter combination
result.best_score_                   # Best cross-validation score
result.best_estimator_                    # Model trained with best parameters
print(result.best_index_)
```

```python
dt = result.cv_results_                     # Detailed results for all combination.
df = pd.DataFrame(dt)
df
```

GridSearchCV with Multiple Estimators :

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC

#providing the multiple eatimators to the gridsearchcv :
grid_param = {"decision_tree":{"model":DecisionTreeClassifier(),"param":{"max_depth":[10,12,15,2],"criterion":["gini","entropy"]}},
              "svc":{"model":SVC(gamma="auto"),"param":{"kernel":["linear","rbf"],"C":[1,10,15]}}
  }
dict1 = []
for key,val in grid_param.items():
  scores = GridSearchCV(val["model"],val["param"],cv=5,return_train_score=False)
  scores.fit(x,y)
  dict1.append({"model":key,
                "best_score":scores.best_score_,
                "best_param":scores.best_params_
  }

  )
dict1
```
